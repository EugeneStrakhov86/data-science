```python
def experience_to_number(text):
  '''
  Перетворення текстового значення досвіду роботи у числове.
  '''
  if text == '3 місяці':
    return 0.25
  first_element = text.split()[0]
  if first_element == 'Пів':
    return 0.5
  elif first_element == 'Менше':
    return 0.15
  elif first_element == '1,5':
    return 1.5
  else:
    return float(first_element)

dou['work_experience'] = dou['work_experience'].map(experience_to_number)
```

## Агрегація та групування

```python
dou.groupby('title')['salary'].median().sort_values(ascending=False)
```

```python
dou.groupby('title')['salary'].median().sort_values(ascending=False).plot(kind='bar')
plt.ylabel('Median salary, $ per month')
plt.title('Median salary by title')
plt.show()
```

```python
median_salary = dou['salary'].describe()['50%']
dou.groupby('title')['salary'].median().sort_values().plot(kind='barh')
plt.axvline(x=median_salary, color='green')
plt.xlabel('Median salary, $ per month')
plt.title('Median salary by title')
plt.show()
```

```python
dou.groupby('title')['salary'].agg(['count', 'median']).sort_values('median', ascending=False)
```

```python
df_by_title = dou.groupby('title')[['age', 'work_experience', 'salary']].agg({'age': np.mean,
                                                                              'work_experience': [np.min, np.mean, np.max],
                                                                              'salary': np.median})
np.round(df_by_title, 1).sort_values(('salary', 'median'), ascending=False)
```

```python
english_order = ['Elementary', 'Pre-Intermediate', 'Intermediate', 'Upper-Intermediate', 'Advanced']
title_order = dou.groupby('title')['salary'].median().sort_values(ascending=False).index
df_pivot = dou.pivot_table(index='title', columns='english_level', values='salary', aggfunc='median')
df_pivot.reindex(english_order, axis=1).reindex(title_order, axis=0)
```
