```r
set.seed(0)
X <- NULL
Z <- rnorm(500)
X[1] = 0
for (i in 2:500) {
    X[i] <- X[i-1] + Z[i]
}
```

```r
auto.arima(X)
```

```r
model.fit <- arima.sim(list(order = c(0,1,0)), sd=sqrt(0.9761), n = 500)
```

```r
X <- ts(X, start=c(1900, 1, 1), frequency=365)
model.fit <- ts(model.fit, start=c(1900, 1, 1), frequency=365)
```

```r
autoplot(X) + autolayer(model.fit)
```
